```python
import pandas as pd
```

```python
YEAR = 2022
```

```python
excel = pd.ExcelFile('cy_%s_raw.xlsx' % YEAR)
excel.sheet_names
```

```python
SHEET_NAME = 'Πληρωμές δικαιούχων FY {}'.format(YEAR) # means 'Beneficiary Payments'
#SHEET_NAME = 'Πληρωμές_Δικαιούχων_FY_2018'
SHEET_NAME = 'Πληρωμές - Payments FY2022'
```

```python
df = pd.read_excel('cy_%s_raw.xlsx' % YEAR, sheet_name=SHEET_NAME, skiprows=1)
original_columns = list(df.columns)
df.head()
```

```python
original_columns
```

```python
translations = [
    'recipient_name',
    'application_code',
    'programme_type',
    'scheme_code_short',
    'EAGF',
    'EAFRD',
    'EAFRD national',
    'total',
    'total + national',
    'recipient_postcode',
    'recipient_location',
    'scheme_code',
    'scheme_description',
]
```

```python
df = df.rename(columns=dict(zip(original_columns, translations)))
df['year'] = YEAR
df = df.drop(columns=['scheme_code'])
df['scheme'] = df.apply(lambda x: '%s - %s' % (x['scheme_code_short'], x['scheme_description']), 1)
df.head()
```

```python
# There only ever is one value filled EAGF or EAFRD
((df[['EAGF', 'EAFRD']].fillna(0.0) == 0.0).sum(1) > 1).any()
```

```python
from slugify import slugify

df['country'] = 'CY'
df['currency'] = 'EUR'

df['recipient_id'] = df.apply(lambda x: 'CY-%s' % x['application_code'] if pd.notnull(x['application_code']) else
                                        'CY-%s-%s' % (x['recipient_postcode'], slugify(
                                            '' if pd.isnull(x['recipient_name']) else x['recipient_name'])), 1)

df['amount'] = df.apply(lambda x: x[['EAGF', 'EAFRD']].sum(), 1)
df = df.drop(['application_code', 'EAFRD', 'EAGF', 'EAFRD national', 'total + national', ], axis=1)
df.head()
```

```python
df.to_csv('cy_%s.csv.gz' % YEAR, index=False, encoding='utf-8', compression='gzip')
```
