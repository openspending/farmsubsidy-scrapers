```python
from glob import glob
import os

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./tamkereso_1000081/export.csv", encoding='latin-1', delimiter=';', converters={'Postal code': str})
df["year"] = 2022
df.head(20)
```

```python
df.columns
```

```python
df = df.rename(columns={
        'Amount of support (Ft)': 'amount',
        'Name of beneficiary': 'recipient_name',
        'Postal code': 'recipient_postcode',
        'Address': 'recipient_address',
        'Municipality': 'recipient_location',
        'Measure': 'scheme_detail'
    })
df['recipient_name'] = df['recipient_name'].apply(lambda x: np.nan if x == 'magánszemély' else x)
df['scheme'] = df.apply(lambda x: '%s (%s, %s)' % (x['scheme_detail'], x['FUND'], x['Resource']), 1)
df.head()
```

```python
df['currency'] = 'HUF'
df['country'] = 'HU'
del df["FUND"]
del df["Resource"]
del df["scheme_detail"]
df.head()
```
